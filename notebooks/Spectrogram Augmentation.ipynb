# Spectrogram Augmentation

We use this notebook to explore augmenting the spectrogram to match training to inference. 

```python
%matplotlib inline

import re
import sys
import random

# Setup the "PYTHONPATH"
sys.path.insert(0, '../')

from src.hparams import set_hparams

set_hparams()
```

```python
import numpy as np

from src.audio import get_log_mel_spectrogram
from src.audio import read_audio
from src.utils import plot_log_mel_spectrogram

clip = '../tests/_test_data/lj_speech_24000.wav'
signal = read_audio(clip)

log_mel_spectrogram, padding = get_log_mel_spectrogram(signal)
print('Shape of spectrogram %s' % str(log_mel_spectrogram.shape))

# Pad signal such that ``padded_signal.shape[0] / log_mel_spectrogram.shape[0] == frame_hop``
padded_signal = np.pad(signal, padding, mode='constant', constant_values=0)
padded_signal = (1 / np.max(np.abs(padded_signal))) * padded_signal

_ = plot_log_mel_spectrogram(log_mel_spectrogram) # ``_=`` to avoid unnecessary visualization  
```

```python
from math import floor
from scipy.ndimage import filters

from torch.nn import MSELoss

import torch

criterion = MSELoss()
window_len = 9
smoothed_log_mel_spectrogram = np.array(log_mel_spectrogram, copy=True)  
for i in range(log_mel_spectrogram.shape[0]):
    fft = log_mel_spectrogram[i]
    window = np.hanning(window_len)
    smoothed_log_mel_spectrogram[i] = filters.convolve1d(fft, window/window.sum())
    
_ = plot_log_mel_spectrogram(log_mel_spectrogram) # ``_=`` to avoid unnecessary visualization  
_ = plot_log_mel_spectrogram(smoothed_log_mel_spectrogram) # ``_=`` to avoid unnecessary visualization  

print('MSE Loss between smoothed and ground truth:')
criterion(torch.from_numpy(smoothed_log_mel_spectrogram), torch.from_numpy(log_mel_spectrogram))
```
